```python
#take text
searchinput = "Buy mobile phone buy phone online"

#conver to lower case
cleantext = searchinput.lower()

# Remove symbols
for symbol in [".", ",", "!", "?", ":"]:
    cleantext = cleantext.replace(symbol, "")

# Split words
wordlist = cleantext.split()

#count words 
wordcount = {}

for word in wordlist:
    if word in wordcount:
        wordcount[word] += 1
    else:
        wordcount[word] = 1

# Filter 
repeatwords = {i: j for i, j in wordcount.items() if j > 1}
print(repeatwords)
```

```python
hourly= [3, 4, 7, 8, 10, 12, 5]

# list
readings = []

# Loop data
for hour, reading in enumerate(hourly):
    
    #cheking is evn
    if reading % 2 == 0:
        readings.append((hour, reading))

# result
print("Valid Sensor Readings (hour, Value):")
print(readings)
```

```python
# Email list
emails = [
    "nithin@gmail.com",
    "ridhi@yahoo.com",
    "madhu@gmail.com",
    "bunny@gmail.com",
    "kiran@yahoo.com"
]

# Domain count
counter = {}

# domain
for email in emails:
    domain= email.split("@")[1]
    
    if domain in counter:
        counter[domain] += 1
    else:
        counter[domain] = 1

# users total 
total= len(emails)

# Printing percent
for domain, count in counter.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")

    
```

```python
# Sales data
salesdaily=[1200, 1500, 900, 2200, 1400, 3000]

# Avg value
avgsales = sum(salesdaily) / len(salesdaily)

# Spike limit
limit =avgsales * 1.30

# Check spikes
for index, sale in enumerate(salesdaily, start=1):
    if sale > limit:
        print(f"Day {index}: {sale}")
```

```python
# User IDs
user_reg =["usera", "userb", "userc", "userc", "userb", "usera"]

# Count IDs
freq_id = {}

for user in user_reg:
    if user in freq_id:
        freq_id[user] += 1
    else:
        freq_id[user] = 1

# Show duplicates
for user, count in freq_id.items():
    if count > 1:
        print(f"{user} → {count} times")
```
