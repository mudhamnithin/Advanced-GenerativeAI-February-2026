Problem Statement 1:Smart Parking Lot Management System

```python
def entry_log(vechile,capacity):
    if(total_vechile(vechile)>=capacity):
        print("capacity exceed")
    else:
        print("vechile entry sucessfull")
        vechile.append("IN")
    print("the total vechiles in the parking is:",total_vechile(vechile))    
def exit_log(vechile):
    if(total_vechile(vechile)==0):
        print("no vehiles in the parking ")
    else:    
        print("vechile exit sucessfull")
        vechile.append("OUT")  
    print("the total vechiles in the parking is:",total_vechile(vechile))
def total_vechile(vechile):
    in_count=0
    out_count=0
    for i in vechile:
        if( i=="IN"):
            in_count+=1;
        else:
            out_count+=1;
    return (in_count-out_count)        
capacity=int(input("enter the capacity :"))
vechile=["IN","OUT","IN","OUT","IN","OUT","IN"]
peak_usage = [total_vechile(vechile)]
entry_log(vechile,capacity)
exit_log(vechile)
total_vechile(vechile)
print("peak number:",peak_usage[0])
```

Problem Statement 2:Online Food Delivery Time Estimator

```python
def estamate_ETA(distance,traffic_level,wheather):
    ETA=0
    ETA=distance*2
    if(traffic_level=="H"):
        ETA=ETA*3
    elif(traffic_level=="M"):
        ETA=ETA*2
    if(wheather=="R"):
        ETA=ETA*1.5
    elif(wheather=="C"):
        ETA=ETA*0.5
    return ETA     
    
distance=int(input("Enter the distance:"))
traffic_level=input("Enter the trafic level H for high,L-for Low ,M-medium:")
wheather=input("enter the weather R for rainy,S for sunny,C for cold:")
print("Estimated Delivery Time:", int(estamate_ETA(distance, traffic_level, wheather)), "minutes")
```

Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

```python
def analyze(book_seats,total_seats):
    booked=0
    not_booked=0
    for i in book_seats:
        if(i==1):
            booked+=1
        else:
            not_booked+=1
    status=(booked*100/total_seats)
    print("Occupancy",status,"%")
    if(status == 100):
        print("Show Status: Housefull")
        print("Suggestion: Open Additional Show")
    elif(status >= 50):
        print("Show Status: greater than half filled")
    else:
        print("Show Status: half filled")
        
total_seats=int(input("Enter the total Number of seats:"))
book_seats=[]
for i in range(total_seats):
    book_seats.append(int(input("enter the seat is booked or not 0 is not 1 is booked")))
analyze(book_seats,total_seats)    
```

Problem Statement 4:Cloud Server Load Classification System

```python
def check_server(CPU_readings):
    avg=sum(CPU_readings)/len(CPU_readings)
    if(avg<50):
        print("server status:Normal",avg,"%")
    elif(avg>=50 and avg<=80):
        print("server status:Warning",avg,"%")   
    else:
        print("server status:Critical",avg,"%")
        

n=int(input("enter the number of inputs:"))
CPU_readings=[]
for i in range(n):
    CPU_readings.append(int(input()))
check_server(CPU_readings)    
    
    
```

Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
def check_resource_usage(resources):
    high_usage = []

    for name, hours in resources.items():
        if hours > 8:
            high_usage.append(name)

    if high_usage:
        print("over used resourse:", ", ".join(high_usage))
        print("energy alert: Yes")
    else:
        print("over used resourse:")
        print("energy alert: no")


resources = {
    "projector": 6,
    "aC": 9,
    "lights": 4
}

check_resource_usage(resources)
```

 Problem Statement 6: Online Event Registration Capacity Controller

```python
def check_status(capacity,reg):
    confirmed = min(reg, capacity)
    waitlisted = max(0, reg - capacity)

    print("Confirmed Registrations:",confirmed)
    print("Waitlisted Users",waitlisted)
    if(reg>capacity):
        print("Registration Status: Closed")
    else: 
        print("Registration Status: open")

capacity=int(input("Enter the event capacity:"))
reg=int(input("Enter the number of registrations received:"))
check_status(capacity,reg)
```
