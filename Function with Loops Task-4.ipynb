Problem 2: Inventory Reorder Alert System

```python
#verify the stock products
def verify_mystock(products,stock):
    if(stock>15):
        #product is there 
        print(products,"Stock Ok")
    else:
        #stock is reorder
        print(products,"Reorder Alert")
# enter the  products and stock        
my_stock={"soaps":34,"toothpaste":3,"brush":76,"shampoo":76,"books":9}
for products,stock in my_stock.items():
    verify_mystock(products,stock)
    
        
    
```

Problem 3: Student Result Processing System

```python
def check_avg(i):
    return (sum(i)/len(i))
#enter the number of students 
n=int(input("enter the number of students:"))
marks=[]
#enter the marks
for i in range(n):
    marks.append(list(map(float,input().split())))
#sum of the marks
for i in marks:
    print("for student is",check_avg(i))
    
    
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def fare_cal(km,p):
    fare=50+km*12
    if p=="peak":
        fare=fare+fare*(25/100)
        return fare
    else :
        return fare 
def retry():
    print("entered in valid km or p")
    
while True:
    km=float(input())
    p=input()
    if(km<0):
        retry()
        continue
    elif p not in ["peak", "nonpeak"]:
        retry()
        continue
    else:
        print(fare_cal(km,p))
        break; 
```
