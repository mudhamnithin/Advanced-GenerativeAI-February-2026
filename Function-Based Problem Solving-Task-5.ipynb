Problem 1: Social Media – Post Engagement Analyzer

```python
def verify_viral(list_likes):
    likes_count=0;
    for i in list_likes:
        likes_count=likes_count+i
        
    print("Total Likes:", likes_count) 
    
    if(likes_count>=1000):
        print("post_status:Viral Post")
    else:
        print("post_status:Normal Engagement")
n=int(input())
list_likes=[]
for i in range(n):
    list_likes.append(int(input()))
verify_viral(list_likes)
    
    
```

```python
def verify_stock(medicine_items):
    for i in medicine_items:
        print("medicine_stock:",i)
        if(i>=10):
            print("Status:sufficient Stock")
        elif(i<10 and i>=0):
            print("Status: Low Stock Alert")
        else:
                print("enter the valid number")
n=int(input())
medicine_items=[]
for i in range(n):
    medicine_items.append(int(input()))
verify_stock(medicine_items)    
```

```python
def verify_rainfall(day_rainfall,n,req):
    summ=0;
    for i in day_rainfall:
        summ=summ+i
    average=summ/n;
    if(average>=req):
        print("Average Railfall:",average)
        print(" Adequate Rainfall")
    else:
        print("Average Railfall:",average)
        print(" InAdequate Rainfall")
day_rainfall=[]
n=int(input("enter the numbeer of days"))
r=int(input("Enter the required"))
for i in range(n):
    day_rainfall.append(float(input()))
verify_rainfall(day_rainfall,n,r)    
    
```

```python
def verify_duplicates(account_usernames,n):
    if(n!=len(account_usernames)):
        print("duplicate found:Yes")
    else:   
        print("duplicate found:NO")
    
account_usernames=set()
n=int(input())
for i in range(n):
    account_usernames.add(input())
verify_duplicates(account_usernames,n)    
    
    
    
    
```

```python
def check_health(health_status):
    for i in health_status:
        print("error count:",i)
        if(i==0):
            print("System status:healthy")
        elif(i<=5):
            print("System status:Minor issues")
        else:    
            print("System status:Critical issues")
    

health_status=[]
n=int(input())
for i in range(n):
    health_status.append(int(input()))
check_health(health_status)    
    
```

Problem 8: Banking – Daily Transaction Limit Checker

```python
def verify_transaction(enter_daily_limit,trans_amount):
    print("Transaction Amount:",trans_amount)
    if(enter_daily_limit>trans_amount):
        print("Transaction Status: Aproved")
    else:
        print("Transaction Status: Rejected")
enter_daily_limit=int(input("enter the daily limit:"))
trans_amount=int(input("enter the amount to do transaction:"))
verify_transaction(enter_daily_limit,trans_amount)
```

```python
def check_eligibility(attendance):
    att=0
    for i in attendance:
      att=att+i
    
    att=att/len(attendance)
    print("Attendance Percentage:",att)
    if(att>=75):
        print("Exam Eligiblility:Eligible")
    else:
        print("Exam Eligibility:Not Eligible")
    
    
    
days=int(input("enter the number of days"))
attendance=[]
for i in range(days):
    attendance.append(float(input()))
check_eligibility(attendance)    
    
```

```python
def calculate_bill(units):
    total_bill=0
    if(units>0 and units<=100):
        total_bill=3*units
    elif(units>=101 and units<=200):
        total_bill=5*units
    else:
        total_bill=7*units
    print("price of bill is:",total_bill)
    if(total_bill<500):
        print("Low usange")
    elif(total_bill>=500 and total_bill<1500):
        print("Moderate usange")  
    else:
        print("High Usage")
        

units=float(input("enter the number of units:"))
calculate_bill(units)
```
