2. Pass / Fail Analyzer


```python
#students marks
marks = [45, 78, 90, 33, 60]
count_pass=0
count_fail=0
for i in marks:
    if i>=50:
        count_pass=count_pass+1
    else:
        count_fail=count_fail+1
print("Number of students passed is",count_pass)
print("Number of students failed is",count_fail)
```

3. Simple Data Cleaner

```python

names=[" Alice ", "bob", " CHARLIE "]
names2=[]
for i in names:
    temp=i.strip().lower()
    names2.append(temp)
print(names2)    
```

4. Message Length Analyzer


```python
messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
    msg=len(i);
    print("Message:",i)
    print("length of Message:",msg)
    if msg>10:
        print("status:flag")
    else:
        print("status:ok")
```

5. Error Message Detector
