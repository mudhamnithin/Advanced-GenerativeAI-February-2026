Problem 1: Unique Words in a Sentence

```python
words=input("enter the sentence:")
individual_words=[]
str1=""
for i in words:
    if(i!=' '):
        
        str1=str1+i;
    else:
        individual_words.append(str1)  
        str1=""

individual_words.append(str1)
unique_words=set()
for i in individual_words:
    unique_words.add(i)
    
print("Number of unique words:",len(unique_words))
print("the words in the sentence are:",unique_words) 
```

Problem 2: Highest Salary from Employee Data

```python
employee_data={"nithin":20000,"bunny":30000,"mani":50000,"rishi":34000}
maximum_salary=0;
highest_salary_emplyee=""
for name,salary in employee_data.items():
    if salary>maximum_salary:
        maximum_salary=salary
        highest_salary_emplyee=name

print("highest salary candidate:",highest_salary_emplyee)
print("highest salary:",maximum_salary)
```

Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]
maxi=float('-inf')
mini=float('inf')
for i in numbers:
    if(i>maxi):
        maxi=i;
    if(i<mini):
        mini=i;
print("maximum Number is ",maxi)
print("Minimum Number is ",mini)
```

Problem 4: Count Products Above a Price Threshold

Problem 5: Calculate Attendance Percentage


```python
attendance = ["P", "P", "A", "P", "P"]
n=len(attendance)
count=0
for i in attendance:
    if i=="P":
        count=count+1
print(" attenance is :",(count/n)*100)        
```
