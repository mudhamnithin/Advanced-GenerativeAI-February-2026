Problem 1: Smart Light Controller

```python
class home:
    def __init__(self,room):
        self.room=room
        self.status="off"
    def light_on(self):
        self.status="on"
    def light_off(self):
        self.status="off"
    def status_light(self):
        print("status of the light in:",self.room,"is",self.status)
room=input("enter the room:")
action=input("enter the action:")
light=home(room)
if(action=="on"):
    light.light_on()
elif(action=="off"):
    light.light_off()
else:
    print("entered invalid action")
light.status_light()    
        
```

Problem 2: Employee ID Card System


```python
class employee_details:
    def __init__(self,name,idd,dept):
        self.name=name
        self.idd=idd
        self.dept=dept
    def display(self):
        print("Employee Id Card")
        print("Name:",self.name)
        print("ID:",self.idd)
        print("Depatment:",self.dept)
n=int(input("enter the nummber of employees"))
for i in range(n):
    name=(input())
    emp_id=(input())
    dept=(input())
    emp=employee_details(name,emp_id,dept)
    emp.display()
        
    
```

Problem 3: Mobile Contact Record

```python
class contacts:
    def __init__(self,name,phone_number):
        self.name=name
        self.phone_number=phone_number
    def display(self):
        print("details")
        print("Contact Name:",self.name)
        print("Phone",self.phone_number)
n=int(input("enter the number of contacs to enter"))                
for i in range(n):
    name=(input())
    phone=int(input())
    person=contacts(name,phone)
    person.display()
```
