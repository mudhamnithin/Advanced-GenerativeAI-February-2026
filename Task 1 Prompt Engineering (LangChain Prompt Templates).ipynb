```python

def chat_prompt(role, topic):
    
    if role == "teacher":
        system_msg = "You are a teacher"
    elif role == "interviewer":
        system_msg = "You are an interviewer"
    elif role == "motivator":
        system_msg = "You motivate people"
    else:
        system_msg = "You are helpful"

    chat = ChatPromptTemplate.from_messages([
        ("system", system_msg),
        ("user", "Explain {topic}")
    ])

    return chat.format_messages(topic=topic)

print(chat_prompt("teacher", "Neural Networks"))
```

```python

def validate(audience, tone):
    
    audiences = ["beginner", "intermediate", "expert"]
    tones = ["formal", "casual", "fun"]

    if audience not in audiences:
        audience = "beginner"

    if tone not in tones:
        tone = "formal"

    return audience, tone

print(validate("kids", "cool"))
```

```python
# Task 6: main function

def generate_prompt(topic, audience, tone, style):
    
    audience, tone = validate(audience, tone)

    if style == "teaching":
        temp = "Explain {topic} for {audience} in {tone} way"
    elif style == "interview":
        temp = "Ask questions on {topic} for {audience}"
    elif style == "storytelling":
        temp = "Explain {topic} as a story for {audience}"
    else:
        temp = "Explain {topic}"

    template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=temp
    )

    return template.format(topic=topic, audience=audience, tone=tone)

# test
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```
