```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
model_name="microsoft/DialoGPT-small"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
chat_history_ids = None
```

```python
print("hi! my name is tara iam your chat box how can i asiist you ...")
```

```python
while True:
    
    user_text = input("You: ")

    if user_text.lower() in ["exit", "quit"]:
        print("Tara: Bye! Have a nice day.")
        break

    encoded_input = tokenizer.encode(user_text + tokenizer.eos_token, return_tensors='pt')

    if chat_history_ids is not None:
        full_input = torch.cat([chat_history_ids, encoded_input], dim=-1)
    else:
        full_input = encoded_input

    chat_history_ids = model.generate(
        full_input,
        max_length=1000,
        pad_token_id=tokenizer.eos_token_id,
        do_sample=True,
        top_k=40,
        top_p=0.9
    )

    tara_reply = tokenizer.decode(
        chat_history_ids[:, full_input.shape[-1]:][0],
        skip_special_tokens=True
    )

    print("Tara:", tara_reply)
```
